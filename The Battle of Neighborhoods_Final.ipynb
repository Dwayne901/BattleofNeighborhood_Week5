## Introduction/Business Problem

## Analyzing the rate of vaccine distribution in the European Union

The COVID-19 pandemic in the 21st century caused a huge impact on both the economic, and the social sides of the World. Major pharmaceutical companies put forward sophisticated designs to create a vaccine against COVID-19, and so far, 6 vaccines are currently being used in the EU. The objective of this paper is to find out the current vaccine distribution rates throughout the European Union to see the number of vaccines distributed among each age group. The data is partitioned into 8 sets, the first being from the ages of 18-24, the second being 25-49, the third being 50-59, the fourth being 60-69, the fifth being 70-79, and the sixth being 80+. We also have separate partitions for healthcare workers and ‘Long Term Care Facility’(LTCF) workers. 

The data set monitors the age group, which week the vaccine was administered, type of vaccine administered, and the uptake percentages for both the first, and second doses, based on the population of the country in question. The different age ranges will show us the percentage of each age range that has gotten each dose of the vaccine, showing us the rate at which the vaccine is delivered in each country within the European Union in the time frame from the end of December 2020 to the 3rd week of February. The data in the dataset is pre-processed and normalized. We used the Frequency Distribution method, where the frequency here is the types of vaccines used in each country and how its distributed throughout the European Union. After we analyse the data, we will be able to recognize patterns describing the % uptake for each brand of vaccine in each age group, and chart graphs and models to represent them.

__Keywords:__ Vaccine, Data Analytics, Covid-19, European Union

## Data

The data set provides an overview of the progress in the rollout of COVID-19 vaccines in adults 
(aged 18 years and above) across the EU. It presents the number of vaccine doses distributed by 
manufacturers to each country for the different vaccines authorized for use in the EU and the 
number of first, second or unspecified doses administered to adult individuals overall, by age 
group and other selected target groups including healthcare workers (HCW) and residents in 
long-term care facilities (LTCF).
The timeframe of our data set is from the end of December 2020 to the 3rd week of February.
The data presented in the data set is collected through The European Surveillance System 
(TESS). EU countries are requested to report basic indicators (number of vaccine doses 
distributed by manufacturers, number of first, second and unspecified doses administered) and 
data by target groups at national level twice a week (every Tuesday and Thursday).

The dataset contains 5 vaccine brands, COM(Comirnaty), MOD(Moderna), AZ(AstraZeneca), 
SPU(Sputnik), UNK(Unknown vaccine types).

The attributes of the data set are:
1. Country
2. Age group
3. Vaccine Brand
4. Week
5. Population of the country
6. 1st Dose and 2nd Dose of the vaccine administered

The data of the vaccine distribution is from https://www.ecdc.europa.eu/en/publicationsdata/covid-19-vaccine-tracker which is a free and open portal where anyone can access the data regarding the vaccine distribution in Europe. The data is available is XLS and CSV format.

# Methodology

• The data set has been preprocessed, normalized and cleaned.

• Prior to the preprocessing and cleaning of the data, there were about 3258 rows.

• After cleaning the data, we are left with 2588 rows by removing the inconsistent data.

• We use the frequency distribution method, where the frequency is the type of vaccines used in each country and how its distributed throughout the European Union. 

• We used tableau as our platform to process the data.

# Conclusion

• We can infer that COM(Comirnaty) vaccine has the highest distribution rate in Europe 
with the majority of vaccines given to HCW(Healthcare Workers).

• The highest average amount of vaccines were given to ages 25-49 with the exception of 
SPU(Sputnik) vaccine which was given to ages 50 and above.
